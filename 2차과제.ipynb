### 결국, 결측치를 지우고 나면 데이터는 처음 생각했던 것처럼 국가별로 방대하게 생각하기에는 데이터가 매우 협소한 편이지만 어쨌든 (국가, 연도, 비율) 데이터가 있는 것이므로, 이를 통해 상관관계 파악가능

```python
# 폭력 널값 전처리
def ZSDrop(df):
    dfYR = df.dropna(axis = 1, how = 'all')
    dfCR = dfYR.dropna(axis = 0, how = 'all')
    return dfCR
```

```python
burnCR = ZSDrop(burnZS)
goesCR = ZSDrop(goesZS)
neglCR = ZSDrop(neglZS)
refuCR = ZSDrop(refuZS)
```
