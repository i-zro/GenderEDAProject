```python
# 연도별 평균을 내어 시각화 해보자
# 1. 연도별 청소년 평균, 여성출산율 그룹화 하기
faYR = faRST.groupby('time')[['SP.ADO.TFRT', 'SP.DYN.TFRT.IN']].agg(\
    **{'adoAV':pd.NamedAgg(column='SP.ADO.TFRT', aggfunc='mean'),\
       'femAV':pd.NamedAgg(column='SP.DYN.TFRT.IN', aggfunc='mean')})\
.reset_index()
```

```python
faYR
```

```python
# 2. 단위를 맞춰주기 위해서 adoAV를 1000 으로 나눠주었습니다. (청소년 1000명당 출산율)
faYRR = faYR.copy()
faYRR[['adoAV']] = faYR[['adoAV']].apply(lambda x : x / 1000)
faYRR
```

```python
# 전체 연도별 - 성인, 청소년의 출산율 그래프 : 전체적으로 비례관계를 나타냄.
fig = plt.figure(figsize=(15,15))
sns.set_style('whitegrid')

area1 = fig.add_subplot(2,1,1)
area1.set_title('female fertility graph')
area2 = fig.add_subplot(2,1,2)
area2.set_title('adolescent fertility graph')

sns.stripplot(x='time', y='femAV', data=faYRR, ax=area1)
sns.stripplot(x='time', y='adoAV', data=faYRR, ax=area2)
plt.show()
```
